```python
fun_list = []
school_list = []
food_list = []
transport_list = []
for some_list, some_json in [(fun_list, fun_json), (school_list, school_json), (food_list, food_json), (transport_list, transport_json)]:
    for venue in some_json['response']['venues']:
        venue_location = venue['location']
        venue_location_address = ''
        if 'address' in venue_location:
            venue_location_address = venue_location['address']
        some_list.append([venue['name'], venue_location_address, venue['location']['lat'], venue['location']['lng']])
```
